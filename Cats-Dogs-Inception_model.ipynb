## Downloading and Configuring Inception model

```python
import wget 
from tensorflow.keras.applications.inception_v3 import InceptionV3
from tensorflow.keras import layers
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras import Model
from tensorflow.keras.preprocessing.image import ImageDataGenerator as idg
import os
import zipfile
```

```python
local_weight_file = 'C:/Users/KUSHAGRA/Desktop/Cats-Dogs/inception_v3_weights_tf_dim_ordering_tf_kernels_notop.h5'

pre_trained_model = InceptionV3(input_shape=(150, 150, 3), 
                                 include_top=False,
                                 weights=None)

pre_trained_model.load_weights(local_weight_file)
for layer in pre_trained_model.layers:
    layer.trainable = False
```

```python
pre_trained_model.summary()
```

```python
last_layer = pre_trained_model.get_layer('mixed7')
print('Last layer output shape: ', last_layer.output_shape)
last_output = last_layer.output
```

## Adding our own Dense layers

```python
x = layers.Flatten()(last_output)
x = layers.Dense(1024, activation='relu')(x)
x = layers.Dropout(0.2)(x)
x = layers.Dense(1, activation='sigmoid')(x)

model = Model(pre_trained_model.input, x)
model.summary()
```

```python
model.compile(optimizer=RMSprop(learning_rate=0.0001),
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

### Preparing Datasets

```python
base_dir = 'C:/Users/KUSHAGRA/Desktop/Cats-Dogs/Dataset'

train_dir = os.path.join(base_dir, 'train')
validation_dir = os.path.join(base_dir, 'validation')

train_dir_cats = os.path.join(train_dir, 'cats')
train_dir_dogs = os.path.join(train_dir, 'dogs')

validation_dir_cats = os.path.join(validation_dir, 'cats')
validation_dir_dogs = os.path.join(validation_dir, 'dogs')

train_datagen = idg(rescale=1.0/255,
                   rotation_range=40,
                   width_shift_range=0.2,
                   height_shift_range=0.2,
                   shear_range=0.2,
                   zoom_range=0.2,
                   horizontal_flip=True,
                   fill_mode='nearest')

test_datagen = idg(rescale=1.0/255)

train_data = train_datagen.flow_from_directory(train_dir, batch_size=20, class_mode='binary', target_size=(150, 150))

validation_data = test_datagen.flow_from_directory(validation_dir, batch_size=20, class_mode='binary', target_size=(150, 150))
```

### Training the model

```python
history = model.fit(train_data, 
                    validation_data=validation_data,
                    steps_per_epoch=100,
                    epochs=20,
                    validation_steps=50,
                    verbose=2)
```

```python
import matplotlib.pyplot as plt
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc))

plt.figure(figsize=(18, 12))
plt.plot(epochs, acc, 'r', label='Training accuracy')
plt.plot(epochs, val_acc, 'b', label='Validation accuracy')
plt.legend('Training and Validation accuracy')
plt.title('Accuracy', size=20)
plt.legend(loc=0)
plt.show()
```

```python
plt.figure(figsize=(18, 12))
plt.plot(epochs, loss, 'r', label='Training Loss')
plt.plot(epochs, val_loss, 'b', label='Validation Loss')
plt.legend('Training and Validation Loss')
plt.title('Loss', size=20)
plt.legend(loc=0)
plt.show()
```
